```python
import requests #web sayfalarını çekmek için
import pandas as pd #veri analiz kütüphanesi
import matplotlib.pyplot as plt #grafik için
from bs4 import BeautifulSoup #veri analizi ve veri çekmek için kütüphane
import random

url = 'https://www.imdb.com/list/ls091520106/'
response = requests.get(url) #web sayfasının içeriğini göstermek için

#film bilgileri
films = []
soup = BeautifulSoup(response.text, 'html.parser') #HTML içeriğini BeautifulSoup kütüphanesi kullanılarak analiz etmek için
for film in soup.find_all('div', class_='lister-item-content'):
    title = film.find('h3', class_='lister-item-header').find('a').text
    year = film.find('h3', class_='lister-item-header').find('span', class_='lister-item-year').text
    rating = film.find('div', class_='ipl-rating-star small').find('span', class_='ipl-rating-star__rating').text
    genre = film.find('span', class_='genre').text.strip()
    votes = film.find('span', attrs={'name': 'nv'})['data-value']
    votes = "{:,.2f}".format(float(votes)).rstrip('0').rstrip('.')
    #oy sayısı float ile düzenli değere dönüşür.ardından ondalık basamakları belirler({:,.2f}ile) -
    # rstrip('0').rstrip('.') ->gereksiz sıfır ve noktaları temizlemek için
    films.append({
        'film_adi': title,
        'yil': year,
        'puan': rating,
        'tür': genre,
        'oylar': votes
    })
turler_film_sayisi = {} #her türün film sayısını saklamak için

for film in films:
    print(film)

    ilk_tur = film['tür'].split(', ')[0] #tür bilgisi almak için

    if ilk_tur in turler_film_sayisi:
        turler_film_sayisi[ilk_tur] += 1
    else:
        turler_film_sayisi[ilk_tur] = 1

```
